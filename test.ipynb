```python
import os
import dotenv
import requests
import pandas as pd


dotenv.load_dotenv(".env")


data = requests.get(f"https://www.alphavantage.co/query?function=TIME_SERIES_DAILY&symbol=IBM&outputsize=full&apikey={os.getenv('TOKEN_2')}").json()
data
```

```python
df = pd.DataFrame(data["Time Series (Daily)"])
df = df.T
df.columns = ["open", "high", "low", "close", "volume"]
df.index.name = "date"
df.to_csv("test_data.csv")
```
